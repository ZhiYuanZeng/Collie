```python
import re

# 初始化一个空列表来存储结果
results = []

# 正则表达式模式匹配
# pattern = re.compile(r'./ds_csv_logs/[^#]+#llmllama2_7b#pruner([^#]+)#memory([^#]+)#.*#eval_len(\d+)-')
metric_pattern = re.compile(r'step,Metric ([^#]+)#ppl\n0,([0-9.]+)')
patterns = {
    "pruner": r"#pruner(?P<pruner>[^#]*)",
    "memory": r"#memory(?P<memory>[^#]*)",
    "chunk": r"#chunk(?P<chunk>\d+)",
    "compressed_chunk": r"#compressed_chunk(?P<compressed_chunk>\d+)",
    "memorysize": r"#memorysize(?P<memorysize>\d+)",
    "eval_len": r"#eval_len(?P<eval_len>\d+)"
}

# 按文件路径行分割数据
entries = data.strip().split('\n./')

# 处理每个条目
for entry in entries:
    entry = './' + entry if not entry.startswith('./') else entry
    lines = entry.strip().split('\n')
    if len(lines) > 1:
        # 提取基本信息
        r = {}

        for name, pattern in patterns.items():
            match = re.search(pattern, lines[0])
            if match:
                r[name] = match.group(name)
            else:
                r[name] = None
        
        # 提取ppl值
        arxiv_ppl = None
        github_ppl = None
        for i in range(1, len(lines) - 1, 2):
            metric_match = metric_pattern.match('\n'.join(lines[i:i+2]))
            if metric_match:
                if metric_match.group(1) == 'github_ppl':
                    github_ppl = float(metric_match.group(2))
                elif metric_match.group(1) == 'arxiv_ppl':
                    arxiv_ppl = float(metric_match.group(2))
        r['github_ppl'] = github_ppl
        r['arxiv_ppl'] = arxiv_ppl
        if 'decremental' in lines[0]:
            r['memory'] = 'decremental_chunk'
        # 创建一个字典并添加到结果列表中
        if github_ppl is not None and arxiv_ppl is not None:
            results.append(r)

# 打印结果列表
print(results)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_ppl(fixed_arxiv_ppl, incremental_arxiv_ppl, decremental_arxiv_ppl, full_attn_arxiv_ppl, memory_sizes, ylabel, figurename):
    plt.figure(figsize=(10, 6))
    min_value = min(fixed_arxiv_ppl + incremental_arxiv_ppl + decremental_arxiv_ppl)
    max_value = max(fixed_arxiv_ppl + incremental_arxiv_ppl + decremental_arxiv_ppl)
    
    if full_attn_arxiv_ppl is not None:
        plt.axhline(y=full_attn_arxiv_ppl, color='black', linestyle='--', linewidth=1, label='Full-Attention PPL')

    x = np.arange(len(memory_sizes))
    width = 0.3
    bars1 = plt.bar(x - width, fixed_arxiv_ppl, width, label='Fixed-size Memory', color=macaron_colors[0])
    bars2 = plt.bar(x, incremental_arxiv_ppl, width, label='Incremental Memory', color=macaron_colors[1])
    bars3 = plt.bar(x + width, decremental_arxiv_ppl, width, label='Decremental Chunk', color=macaron_colors[2])
    
    plt.xlabel('Memory size')
    plt.ylabel(ylabel)
    plt.xticks(x, memory_sizes)
    plt.ylim(min_value - 0.1, max_value + 0.1)
    plt.legend()
    plt.gca().spines['top'].set_visible(False)
    # Add numerical labels on the bars
    def add_labels(bars):
        for bar in bars:
            height = bar.get_height()
            plt.annotate(f'{height:.3f}',
                         xy=(bar.get_x() + bar.get_width() / 2, height),
                         xytext=(0, 3),  # 3 points vertical offset
                         textcoords="offset points",
                         ha='center', va='bottom', fontsize=14)
    
    add_labels(bars1)
    add_labels(bars2)
    add_labels(bars3)
    
    plt.show()
    plt.close()

def visualize_ppl(ppl_data, title, full_attn_arxiv_ppl=None, full_attn_github_ppl=None):
    plt.figure(figsize=(10, 6))
    ppl_data = sorted(ppl_data, key=lambda x: int(x['memorysize']))
    memory_sizes = sorted(list(set([int(d['memorysize']) for d in ppl_data])))

    fixed_arxiv_ppl = [d['arxiv_ppl'] for d in ppl_data if d['memory'] == 'Fixed-Size Memory' and d['arxiv_ppl'] is not None]
    incremental_arxiv_ppl = [d['arxiv_ppl'] for d in ppl_data if d['memory'] == 'Incremental Memory' and d['arxiv_ppl'] is not None]
    decremental_arxiv_ppl = [d['arxiv_ppl'] for d in ppl_data if d['memory'] == 'Decremental Chunk' and d['arxiv_ppl'] is not None]

    print(fixed_arxiv_ppl)
    print(incremental_arxiv_ppl)
    print(decremental_arxiv_ppl)

    plot_ppl(fixed_arxiv_ppl, incremental_arxiv_ppl, decremental_arxiv_ppl, full_attn_arxiv_ppl, memory_sizes, title, 'Arixv PPL')

    incremental_github_ppl = [d['github_ppl'] for d in ppl_data if d['memory'] == 'Incremental Memory' and d['github_ppl'] is not None]
    fixed_github_ppl = [d['github_ppl'] for d in ppl_data if d['memory'] == 'Fixed-Size Memory' and d['github_ppl'] is not None]
    decremental_github_ppl = [d['github_ppl'] for d in ppl_data if d['memory'] == 'Decremental Chunk' and d['github_ppl'] is not None]

    print(incremental_github_ppl)
    print(fixed_github_ppl)
    print(decremental_github_ppl)
    
    plot_ppl(incremental_github_ppl, fixed_github_ppl, decremental_github_ppl, full_attn_arxiv_ppl, memory_sizes, title, 'Github PPL')
```
