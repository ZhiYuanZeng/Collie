```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def visualize_matrices(data1, data2):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))  # 创建两个子图

    # 绘制第一个矩阵
    sns.heatmap(data1, cmap="Blues", ax=ax1, cbar=False)
    xticks1 = np.arange(0, data1.shape[1] + 1, max(1, 4))  # 动态设置刻度
    yticks1 = np.arange(0, data1.shape[0] + 1, max(1, 4))  # 动态设置刻度
    ax1.set_xticks(xticks1)
    ax1.set_yticks(yticks1)
    ax1.set_xticklabels(xticks1)
    ax1.set_yticklabels(yticks1)
    ax1.set_xlabel('Chunk')
    ax1.set_ylabel('Layer')
    ax1.invert_yaxis()
    ax1.set_title('LLama2-7b')

    # 绘制第二个矩阵
    sns.heatmap(data2, cmap="Blues", ax=ax2, cbar=False)
    xticks2 = np.arange(0, data2.shape[1] + 1, 4)  # 动态设置刻度
    yticks2 = np.arange(0, data2.shape[0] + 1, 4)  # 动态设置刻度
    ax2.set_xticks(xticks2)
    ax2.set_yticks(yticks2)
    ax2.set_xticklabels(xticks2)
    ax2.set_yticklabels(yticks2)
    ax2.set_xlabel('Chunk')
    ax2.set_ylabel('Layer')
    ax2.invert_yaxis()
    ax2.set_title('Internlm2-7b')

    # 在两个子图中间添加共享颜色条
    cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])  # 颜色条的位置 [左, 下, 宽, 高]
    norm = plt.Normalize(np.min([data1, data2]), np.max([data1, data2]))
    sm = plt.cm.ScalarMappable(cmap="Blues", norm=norm)
    sm.set_array([])
    fig.colorbar(sm, cax=cbar_ax)

    # 调整布局
    plt.tight_layout(rect=[0, 0, 0.9, 1])
    plt.savefig('./imgs/layer-map.pdf', format='pdf')
    plt.show()
    plt.close()


# 调用函数
visualize_matrices(loaded_data_llama['conv'][:,:-1], loaded_data_internlm['conv'][:,:-1])
```

```python
def visual_vector(data_to_plot, figure_name, model_name):
    pruner_name_transform = {
        'conv': 'SnapKV',
        'streaming_llm': 'StreamingLLM'
    }
    
    # 创建x轴上的取值
    plt.figure(figsize=(8, 6))
    plt.grid(True)
    
    # 设置柱状图的宽度
    bar_width = 1.0
    
    # 计算x轴上的位置
    x = np.arange(len(next(iter(data_to_plot.values()))))
    
    # 绘制双栏柱状图
    for i, (k, v) in enumerate(data_to_plot.items()):
        plt.bar(x + i * bar_width, v, bar_width, label=pruner_name_transform[k], color=macaron_colors[i])
    
    plt.xlabel('Chunk')
    plt.ylabel('Proportion')
    plt.legend(loc='upper left')
    # plt.title(f'Memory Distributions of {model_name}')
    plt.savefig(f'imgs/{figure_name}', format='pdf', bbox_inches='tight')
    plt.show()
    plt.close()
```
