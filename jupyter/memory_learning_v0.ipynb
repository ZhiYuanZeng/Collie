```python
import re
import matplotlib.pyplot as plt

def plot_ppl_vs_step(data: str):
    """
    解析输入字符串中的 step 和 ppl 数据，并绘制折线图。
    
    参数:
    data (str): 包含 step 和 ppl 数据的多行字符串
    
    返回:
    None
    """
    steps = []
    ppls = []

    # 按行划分字符串
    lines = data.strip().split('\n')

    # 正则表达式模式
    pattern = r"step:\s*(\d+)\s*,\s*ppl:\s*([\d.]+)"

    # 逐行读取并解析数据
    for line in lines:
        print(f"Processing line: '{line}'")  # Debug: Print each line
        match = re.match(pattern, line)
        if match:
            step = int(match.group(1))
            ppl = float(match.group(2))
            steps.append(step)
            ppls.append(ppl)
        else:
            print(f"No match for line: '{line}'")  # Debug: Print non-matching lines

    if not steps or not ppls:
        print("No valid data found.")
        return
    
    # 绘制折线图
    plt.figure(figsize=(10, 5))
    plt.plot(steps, ppls, marker='o', linestyle='-', color='b')
    plt.title('Step vs Perplexity')
    plt.xlabel('Step')
    plt.ylabel('Perplexity')
    plt.grid(True)
    plt.show()

# 示例数据
data = """
step: 0, ppl:7.625
step: 1, ppl:8.125
step: 2, ppl:12.1875
step: 3, ppl:8.75
step: 4, ppl:9.625
"""

# 调用函数
plot_ppl_vs_step(data)
```

```python
import re
import matplotlib.pyplot as plt
import numpy as np

def plot_loss_groups(data: str, num_groups: int, title):
    """
    解析输入字符串中的 step 和 loss 数据，将数据分成指定数量的组，并绘制柱状图。
    
    参数:
    data (str): 包含 step 和 loss 数据的多行字符串
    num_groups (int): 要分成的组数量
    
    返回:
    None
    """
    steps = []
    losss = []

    # 按行划分字符串
    lines = data.strip().split('\n')

    # 正则表达式模式
    pattern = r"step:\s*(\d+)\s*,\s*loss:\s*([\d.]+)"

    # 逐行读取并解析数据
    for line in lines:
        match = re.match(pattern, line)
        if match:
            step = int(match.group(1))
            loss = float(match.group(2))
            steps.append(step)
            losss.append(loss)

    if not losss:
        print("No valid data found.")
        return

    # 将数据分成 num_groups 组
    losss_array = np.array(losss)
    num_data_points = len(losss_array)
    
    if num_data_points < num_groups:
        print("Number of groups exceeds the number of data points.")
        return
    # 计算每组的平均值
    indices = np.arange(num_data_points)
    group_indices = np.array_split(indices, num_groups)
    avg_losss = [np.mean(losss_array[group]) for group in group_indices]
    y_min, y_max = min(avg_losss), max(avg_losss)

    print(avg_losss)
    # 绘制柱状图
    plt.figure(figsize=(10, 5))
    plt.bar(range(num_groups), avg_losss, color='b', tick_label=[f'{i+1}' for i in range(num_groups)])
    plt.ylim(y_min - 0.1, y_max + 0.1)
    plt.title(f'Average Loss by Epoch ({title})')
    plt.xlabel('Epoch')
    plt.ylabel('Average Perplexity')
    plt.grid(axis='y')
    plt.show()
```
