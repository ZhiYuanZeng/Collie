```python
import matplotlib.pyplot as plt
import numpy as np

def visual_vector(vector, title):
    plt.bar(range(len(vector)), vector)

    plt.title('Vector Visualization as Bar Chart')
    plt.xlabel('Layer')
    plt.ylabel('Rank')
    plt.title(title)
    plt.show()
```

```python
visual_vector(states[:-1, 1:, 0].diagonal().numpy(), 'rank of residual of keys')
```

```python
visual_vector(states[:-1, 1:, 1].diagonal().numpy(), 'rank of residual of values')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap, Normalize

# 定义颜色和对应的数值
colors = [(0, "blue"), (5, "white"), (10, "red")]  # 0 -> blue, 5 -> white, 10 -> red

# 将数值归一化到 0 到 1 的范围
norm = Normalize(vmin=0, vmax=10)

# 创建自定义的LinearSegmentedColormap
cmap = LinearSegmentedColormap.from_list("custom_cmap", [(norm(v), c) for v, c in colors])

# 创建示例数据
data = np.full((10, 10), 0)

# 绘制图像并使用自定义的颜色映射
plt.imshow(data, cmap=cmap, norm=norm)
plt.colorbar()
plt.show()
```
